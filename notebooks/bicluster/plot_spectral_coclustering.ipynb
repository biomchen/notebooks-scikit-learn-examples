A demo of the Spectral Co-Clustering algorithm<br>
==============================================<br>
This example demonstrates how to generate a dataset and bicluster it<br>
using the Spectral Co-Clustering algorithm.<br>
The dataset is generated using the ``make_biclusters`` function, which<br>
creates a matrix of small values and implants bicluster with large<br>
values. The rows and columns are then shuffled and passed to the<br>
Spectral Co-Clustering algorithm. Rearranging the shuffled matrix to<br>
make biclusters contiguous shows how accurately the algorithm found<br>
the biclusters.<br>

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from sklearn.datasets import make_biclusters
from sklearn.cluster import SpectralCoclustering
from sklearn.metrics import consensus_score
```

```python
data, rows, columns = make_biclusters(
    shape=(300, 300), n_clusters=5, noise=5,
    shuffle=False, random_state=0)
```

```python
plt.matshow(data, cmap=plt.cm.Blues)
plt.title("Original dataset")
```

shuffle clusters

```python
rng = np.random.RandomState(0)
row_idx = rng.permutation(data.shape[0])
col_idx = rng.permutation(data.shape[1])
data = data[row_idx][:, col_idx]
```

```python
plt.matshow(data, cmap=plt.cm.Blues)
plt.title("Shuffled dataset")
```

```python
model = SpectralCoclustering(n_clusters=5, random_state=0)
model.fit(data)
score = consensus_score(model.biclusters_,
                        (rows[:, row_idx], columns[:, col_idx]))
```

```python
print("consensus score: {:.3f}".format(score))
```

```python
fit_data = data[np.argsort(model.row_labels_)]
fit_data = fit_data[:, np.argsort(model.column_labels_)]
```

```python
plt.matshow(fit_data, cmap=plt.cm.Blues)
plt.title("After biclustering; rearranged to show biclusters")
```
