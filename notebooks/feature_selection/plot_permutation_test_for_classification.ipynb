Test with permutations the significance of a classification score<br>
=================================================================<br>
In order to test if a classification score is significative a technique<br>
in repeating the classification procedure after randomizing, permuting,<br>
the labels. The p-value is then given by the percentage of runs for<br>
which the score obtained is greater than the classification score<br>
obtained in the first place.<br>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import permutation_test_score
from sklearn import datasets
```

#############################################################################<br>
Loading a dataset

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
n_classes = np.unique(y).size
```

Some noisy data not correlated

```python
random = np.random.RandomState(seed=0)
E = random.normal(size=(len(X), 2200))
```

Add noisy data to the informative features for make the task harder

```python
X = np.c_[X, E]
```

```python
svm = SVC(kernel='linear')
cv = StratifiedKFold(2)
```

```python
score, permutation_scores, pvalue = permutation_test_score(
    svm, X, y, scoring="accuracy", cv=cv, n_permutations=100, n_jobs=1)
```

```python
print("Classification score %s (pvalue : %s)" % (score, pvalue))
```

#############################################################################<br>
View histogram of permutation scores

```python
plt.hist(permutation_scores, 20, label='Permutation scores',
         edgecolor='black')
ylim = plt.ylim()
# BUG: vlines(..., linestyle='--') fails on older versions of matplotlib
# plt.vlines(score, ylim[0], ylim[1], linestyle='--',
#          color='g', linewidth=3, label='Classification Score'
#          ' (pvalue %s)' % pvalue)
# plt.vlines(1.0 / n_classes, ylim[0], ylim[1], linestyle='--',
#          color='k', linewidth=3, label='Luck')
plt.plot(2 * [score], ylim, '--g', linewidth=3,
         label='Classification Score'
         ' (pvalue %s)' % pvalue)
plt.plot(2 * [1. / n_classes], ylim, '--k', linewidth=3, label='Luck')
```

```python
plt.ylim(ylim)
plt.legend()
plt.xlabel('Score')
plt.show()
```
