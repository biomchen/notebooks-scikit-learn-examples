Comparison of F-test and mutual information<br>
===========================================<br>
This example illustrates the differences between univariate F-test statistics<br>
and mutual information.<br>
We consider 3 features x_1, x_2, x_3 distributed uniformly over [0, 1], the<br>
target depends on them as follows:<br>
y = x_1 + sin(6 * pi * x_2) + 0.1 * N(0, 1), that is the third features is<br>
completely irrelevant.<br>
The code below plots the dependency of y against individual x_i and normalized<br>
values of univariate F-tests statistics and mutual information.<br>
As F-test captures only linear dependency, it rates x_1 as the most<br>
discriminative feature. On the other hand, mutual information can capture any<br>
kind of dependency between variables and it rates x_2 as the most<br>
discriminative feature, which probably agrees better with our intuitive<br>
perception for this example. Both methods correctly marks x_3 as irrelevant.<br>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_selection import f_regression, mutual_info_regression
```

```python
np.random.seed(0)
X = np.random.rand(1000, 3)
y = X[:, 0] + np.sin(6 * np.pi * X[:, 1]) + 0.1 * np.random.randn(1000)
```

```python
f_test, _ = f_regression(X, y)
f_test /= np.max(f_test)
```

```python
mi = mutual_info_regression(X, y)
mi /= np.max(mi)
```

```python
plt.figure(figsize=(15, 5))
for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.scatter(X[:, i], y, edgecolor='black', s=20)
    plt.xlabel("$x_{}$".format(i + 1), fontsize=14)
    if i == 0:
        plt.ylabel("$y$", fontsize=14)
    plt.title("F-test={:.2f}, MI={:.2f}".format(f_test[i], mi[i]),
              fontsize=16)
plt.show()
```
