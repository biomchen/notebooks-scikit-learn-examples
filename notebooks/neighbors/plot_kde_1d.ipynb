===================================<br>
Simple 1D Kernel Density Estimation<br>
===================================<br>
This example uses the :class:`sklearn.neighbors.KernelDensity` class to<br>
demonstrate the principles of Kernel Density Estimation in one dimension.<br>
The first plot shows one of the problems with using histograms to visualize<br>
the density of points in 1D. Intuitively, a histogram can be thought of as a<br>
scheme in which a unit "block" is stacked above each point on a regular grid.<br>
As the top two panels show, however, the choice of gridding for these blocks<br>
can lead to wildly divergent ideas about the underlying shape of the density<br>
distribution.  If we instead center each block on the point it represents, we<br>
get the estimate shown in the bottom left panel.  This is a kernel density<br>
estimation with a "top hat" kernel.  This idea can be generalized to other<br>
kernel shapes: the bottom-right panel of the first figure shows a Gaussian<br>
kernel density estimate over the same distribution.<br>
Scikit-learn implements efficient kernel density estimation using either<br>
a Ball Tree or KD Tree structure, through the<br>
:class:`sklearn.neighbors.KernelDensity` estimator.  The available kernels<br>
are shown in the second figure of this example.<br>
The third figure compares kernel density estimates for a distribution of 100<br>
samples in 1 dimension.  Though this example uses 1D distributions, kernel<br>
density estimation is easily and efficiently extensible to higher dimensions<br>
as well.<br>

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from distutils.version import LooseVersion
from scipy.stats import norm
from sklearn.neighbors import KernelDensity
```

`normed` is being deprecated in favor of `density` in histograms

```python
if LooseVersion(matplotlib.__version__) >= '2.1':
    density_param = {'density': True}
else:
    density_param = {'normed': True}
```

----------------------------------------------------------------------<br>
Plot the progression of histograms to kernels

```python
np.random.seed(1)
N = 20
X = np.concatenate((np.random.normal(0, 1, int(0.3 * N)),
                    np.random.normal(5, 1, int(0.7 * N))))[:, np.newaxis]
X_plot = np.linspace(-5, 10, 1000)[:, np.newaxis]
bins = np.linspace(-5, 10, 10)
```

```python
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)
fig.subplots_adjust(hspace=0.05, wspace=0.05)
```

histogram 1

```python
ax[0, 0].hist(X[:, 0], bins=bins, fc='#AAAAFF', **density_param)
ax[0, 0].text(-3.5, 0.31, "Histogram")
```

histogram 2

```python
ax[0, 1].hist(X[:, 0], bins=bins + 0.75, fc='#AAAAFF', **density_param)
ax[0, 1].text(-3.5, 0.31, "Histogram, bins shifted")
```

tophat KDE

```python
kde = KernelDensity(kernel='tophat', bandwidth=0.75).fit(X)
log_dens = kde.score_samples(X_plot)
ax[1, 0].fill(X_plot[:, 0], np.exp(log_dens), fc='#AAAAFF')
ax[1, 0].text(-3.5, 0.31, "Tophat Kernel Density")
```

Gaussian KDE

```python
kde = KernelDensity(kernel='gaussian', bandwidth=0.75).fit(X)
log_dens = kde.score_samples(X_plot)
ax[1, 1].fill(X_plot[:, 0], np.exp(log_dens), fc='#AAAAFF')
ax[1, 1].text(-3.5, 0.31, "Gaussian Kernel Density")
```

```python
for axi in ax.ravel():
    axi.plot(X[:, 0], np.full(X.shape[0], -0.01), '+k')
    axi.set_xlim(-4, 9)
    axi.set_ylim(-0.02, 0.34)
```

```python
for axi in ax[:, 0]:
    axi.set_ylabel('Normalized Density')
```

```python
for axi in ax[1, :]:
    axi.set_xlabel('x')
```

----------------------------------------------------------------------<br>
Plot all available kernels

```python
X_plot = np.linspace(-6, 6, 1000)[:, None]
X_src = np.zeros((1, 1))
```

```python
fig, ax = plt.subplots(2, 3, sharex=True, sharey=True)
fig.subplots_adjust(left=0.05, right=0.95, hspace=0.05, wspace=0.05)
```

```python
def format_func(x, loc):
    if x == 0:
        return '0'
    elif x == 1:
        return 'h'
    elif x == -1:
        return '-h'
    else:
        return '%ih' % x
```

```python
for i, kernel in enumerate(['gaussian', 'tophat', 'epanechnikov',
                            'exponential', 'linear', 'cosine']):
    axi = ax.ravel()[i]
    log_dens = KernelDensity(kernel=kernel).fit(X_src).score_samples(X_plot)
    axi.fill(X_plot[:, 0], np.exp(log_dens), '-k', fc='#AAAAFF')
    axi.text(-2.6, 0.95, kernel)
    axi.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
    axi.xaxis.set_major_locator(plt.MultipleLocator(1))
    axi.yaxis.set_major_locator(plt.NullLocator())
    axi.set_ylim(0, 1.05)
    axi.set_xlim(-2.9, 2.9)
```

```python
ax[0, 1].set_title('Available Kernels')
```

----------------------------------------------------------------------<br>
Plot a 1D density example

```python
N = 100
np.random.seed(1)
X = np.concatenate((np.random.normal(0, 1, int(0.3 * N)),
                    np.random.normal(5, 1, int(0.7 * N))))[:, np.newaxis]
```

```python
X_plot = np.linspace(-5, 10, 1000)[:, np.newaxis]
```

```python
true_dens = (0.3 * norm(0, 1).pdf(X_plot[:, 0])
             + 0.7 * norm(5, 1).pdf(X_plot[:, 0]))
```

```python
fig, ax = plt.subplots()
ax.fill(X_plot[:, 0], true_dens, fc='black', alpha=0.2,
        label='input distribution')
colors = ['navy', 'cornflowerblue', 'darkorange']
kernels = ['gaussian', 'tophat', 'epanechnikov']
lw = 2
```

```python
for color, kernel in zip(colors, kernels):
    kde = KernelDensity(kernel=kernel, bandwidth=0.5).fit(X)
    log_dens = kde.score_samples(X_plot)
    ax.plot(X_plot[:, 0], np.exp(log_dens), color=color, lw=lw,
            linestyle='-', label="kernel = '{0}'".format(kernel))
```

```python
ax.text(6, 0.38, "N={0} points".format(N))
```

```python
ax.legend(loc='upper left')
ax.plot(X[:, 0], -0.005 - 0.01 * np.random.random(X.shape[0]), '+k')
```

```python
ax.set_xlim(-4, 9)
ax.set_ylim(-0.02, 0.4)
plt.show()
```
