Decision Tree Regression with AdaBoost<br>
======================================<br>
A decision tree is boosted using the AdaBoost.R2 [1]_ algorithm on a 1D<br>
sinusoidal dataset with a small amount of Gaussian noise.<br>
299 boosts (300 decision trees) is compared with a single decision tree<br>
regressor. As the number of boosts is increased the regressor can fit more<br>
detail.<br>
.. [1] H. Drucker, "Improving Regressors using Boosting Techniques", 1997.<br>

importing necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
```

Create the dataset

```python
rng = np.random.RandomState(1)
X = np.linspace(0, 6, 100)[:, np.newaxis]
y = np.sin(X).ravel() + np.sin(6 * X).ravel() + rng.normal(0, 0.1, X.shape[0])
```

Fit regression model

```python
regr_1 = DecisionTreeRegressor(max_depth=4)
```

```python
regr_2 = AdaBoostRegressor(DecisionTreeRegressor(max_depth=4),
                          n_estimators=300, random_state=rng)
```

```python
regr_1.fit(X, y)
regr_2.fit(X, y)
```

Predict

```python
y_1 = regr_1.predict(X)
y_2 = regr_2.predict(X)
```

Plot the results

```python
plt.figure()
plt.scatter(X, y, c="k", label="training samples")
plt.plot(X, y_1, c="g", label="n_estimators=1", linewidth=2)
plt.plot(X, y_2, c="r", label="n_estimators=300", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Boosted Decision Tree Regression")
plt.legend()
plt.show()
```
