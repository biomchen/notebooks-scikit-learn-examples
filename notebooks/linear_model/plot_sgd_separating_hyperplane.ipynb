SGD: Maximum margin separating hyperplane<br>
=========================================<br>
Plot the maximum margin separating hyperplane within a two-class<br>
separable dataset using a linear Support Vector Machines classifier<br>
trained using SGD.<br>


```python
print(__doc__)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDClassifier
from sklearn.datasets import make_blobs
```

we create 50 separable points

```python
X, Y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=0.60)
```

fit the model

```python
clf = SGDClassifier(loss="hinge", alpha=0.01, max_iter=200)
```

```python
clf.fit(X, Y)
```

plot the line, the points, and the nearest vectors to the plane

```python
xx = np.linspace(-1, 5, 10)
yy = np.linspace(-1, 5, 10)
```

```python
X1, X2 = np.meshgrid(xx, yy)
Z = np.empty(X1.shape)
for (i, j), val in np.ndenumerate(X1):
    x1 = val
    x2 = X2[i, j]
    p = clf.decision_function([[x1, x2]])
    Z[i, j] = p[0]
levels = [-1.0, 0.0, 1.0]
linestyles = ['dashed', 'solid', 'dashed']
colors = 'k'
plt.contour(X1, X2, Z, levels, colors=colors, linestyles=linestyles)
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Paired,
            edgecolor='black', s=20)
```

```python
plt.axis('tight')
plt.show()
```
