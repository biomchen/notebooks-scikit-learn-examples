Sparsity Example: Fitting only features 1  and 2<br>
=========================================================<br>
Features 1 and 2 of the diabetes-dataset are fitted and<br>
plotted below. It illustrates that although feature 2<br>
has a strong coefficient on the full model, it does not<br>
give us much regarding `y` when compared to just feature 1<br>

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
from sklearn import datasets, linear_model
```

```python
X, y = datasets.load_diabetes(return_X_y=True)
indices = (0, 1)
```

```python
X_train = X[:-20, indices]
X_test = X[-20:, indices]
y_train = y[:-20]
y_test = y[-20:]
```

```python
ols = linear_model.LinearRegression()
ols.fit(X_train, y_train)
```

#############################################################################<br>
Plot the figure

```python
def plot_figs(fig_num, elev, azim, X_train, clf):
    fig = plt.figure(fig_num, figsize=(4, 3))
    plt.clf()
    ax = Axes3D(fig, elev=elev, azim=azim)
    ax.scatter(X_train[:, 0], X_train[:, 1], y_train, c='k', marker='+')
    ax.plot_surface(np.array([[-.1, -.1], [.15, .15]]),
                    np.array([[-.1, .15], [-.1, .15]]),
                    clf.predict(np.array([[-.1, -.1, .15, .15],
                                          [-.1, .15, -.1, .15]]).T
                                ).reshape((2, 2)),
                    alpha=.5)
    ax.set_xlabel('X_1')
    ax.set_ylabel('X_2')
    ax.set_zlabel('Y')
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
```

Generate the three different figures from different views

```python
elev = 43.5
azim = -110
plot_figs(1, elev, azim, X_train, ols)
```

```python
elev = -.5
azim = 0
plot_figs(2, elev, azim, X_train, ols)
```

```python
elev = -.5
azim = 90
plot_figs(3, elev, azim, X_train, ols)
```
