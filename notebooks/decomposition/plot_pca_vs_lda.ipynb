Comparison of LDA and PCA 2D projection of Iris dataset<br>
=======================================================<br>
The Iris dataset represents 3 kind of Iris flowers (Setosa, Versicolour<br>
and Virginica) with 4 attributes: sepal length, sepal width, petal length<br>
and petal width.<br>
Principal Component Analysis (PCA) applied to this data identifies the<br>
combination of attributes (principal components, or directions in the<br>
feature space) that account for the most variance in the data. Here we<br>
plot the different samples on the 2 first principal components.<br>
Linear Discriminant Analysis (LDA) tries to identify attributes that<br>
account for the most variance *between classes*. In particular,<br>
LDA, in contrast to PCA, is a supervised method, using known class labels.<br>


```python
print(__doc__)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)
```

Percentage of variance explained for each components

```python
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))
```

```python
plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2
```

```python
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')
```

```python
plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], alpha=.8, color=color,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of IRIS dataset')
```

```python
plt.show()
```
