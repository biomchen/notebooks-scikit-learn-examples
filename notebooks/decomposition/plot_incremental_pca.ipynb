Incremental PCA<br>
===============<br>
Incremental principal component analysis (IPCA) is typically used as a<br>
replacement for principal component analysis (PCA) when the dataset to be<br>
decomposed is too large to fit in memory. IPCA builds a low-rank approximation<br>
for the input data using an amount of memory which is independent of the<br>
number of input data samples. It is still dependent on the input data features,<br>
but changing the batch size allows for control of memory usage.<br>
This example serves as a visual check that IPCA is able to find a similar<br>
projection of the data to PCA (to a sign flip), while only processing a<br>
few samples at a time. This can be considered a "toy example", as IPCA is<br>
intended for large datasets which do not fit in main memory, requiring<br>
incremental approaches.<br>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA, IncrementalPCA
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
n_components = 2
ipca = IncrementalPCA(n_components=n_components, batch_size=10)
X_ipca = ipca.fit_transform(X)
```

```python
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X)
```

```python
colors = ['navy', 'turquoise', 'darkorange']
```

```python
for X_transformed, title in [(X_ipca, "Incremental PCA"), (X_pca, "PCA")]:
    plt.figure(figsize=(8, 8))
    for color, i, target_name in zip(colors, [0, 1, 2], iris.target_names):
        plt.scatter(X_transformed[y == i, 0], X_transformed[y == i, 1],
                    color=color, lw=2, label=target_name)
    if "Incremental" in title:
        err = np.abs(np.abs(X_pca) - np.abs(X_ipca)).mean()
        plt.title(title + " of iris dataset\nMean absolute unsigned error "
                  "%.6f" % err)
    else:
        plt.title(title + " of iris dataset")
    plt.legend(loc="best", shadow=False, scatterpoints=1)
    plt.axis([-4, 4, -1.5, 1.5])
```

```python
plt.show()
```
