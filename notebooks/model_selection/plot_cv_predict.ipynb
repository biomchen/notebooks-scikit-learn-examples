====================================<br>
Plotting Cross-Validated Predictions<br>
====================================<br>
This example shows how to use<br>
:func:`~sklearn.model_selection.cross_val_predict` to visualize prediction<br>
errors.<br>


```python
from sklearn import datasets
from sklearn.model_selection import cross_val_predict
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
lr = linear_model.LinearRegression()
X, y = datasets.load_boston(return_X_y=True)
```

cross_val_predict returns an array of the same size as `y` where each entry<br>
is a prediction obtained by cross validation:

```python
predicted = cross_val_predict(lr, X, y, cv=10)
```

```python
fig, ax = plt.subplots()
ax.scatter(y, predicted, edgecolors=(0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
