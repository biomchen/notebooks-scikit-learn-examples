Decision Tree Regression<br>
===================================================================<br>
A 1D regression with decision tree.<br>
The :ref:`decision trees <tree>` is<br>
used to fit a sine curve with addition noisy observation. As a result, it<br>
learns local linear regressions approximating the sine curve.<br>
We can see that if the maximum depth of the tree (controlled by the<br>
`max_depth` parameter) is set too high, the decision trees learn too fine<br>
details of the training data and learn from the noise, i.e. they overfit.<br>


```python
print(__doc__)
```

Import the necessary modules and libraries

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
```

Create a random dataset

```python
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))
```

Fit regression model

```python
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(X, y)
regr_2.fit(X, y)
```

Predict

```python
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
```

Plot the results

```python
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black",
            c="darkorange", label="data")
plt.plot(X_test, y_1, color="cornflowerblue",
         label="max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```
