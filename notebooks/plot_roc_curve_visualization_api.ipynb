ROC Curve with Visualization API<br>
================================<br>
Scikit-learn defines a simple API for creating visualizations for machine<br>
learning. The key features of this API is to allow for quick plotting and<br>
visual adjustments without recalculation. In this example, we will demonstrate<br>
how to use the visualization API by comparing ROC curves.<br>


```python
print(__doc__)
```

############################################################################<br>
Load Data and Train a SVC<br>
-------------------------<br>
First, we load the wine dataset and convert it to a binary classification<br>
problem. Then, we train a support vector classifier on a training dataset.

```python
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import plot_roc_curve
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
X, y = load_wine(return_X_y=True)
y = y == 2
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
svc = SVC(random_state=42)
svc.fit(X_train, y_train)
```

############################################################################<br>
Plotting the ROC Curve<br>
----------------------<br>
Next, we plot the ROC curve with a single call to<br>
:func:`sklearn.metrics.plot_roc_curve`. The returned `svc_disp` object allows<br>
us to continue using the already computed ROC curve for the SVC in future<br>
plots.

```python
svc_disp = plot_roc_curve(svc, X_test, y_test)
plt.show()
```

############################################################################<br>
Training a Random Forest and Plotting the ROC Curve<br>
--------------------------------------------------------<br>
We train a random forest classifier and create a plot comparing it to the SVC<br>
ROC curve. Notice how `svc_disp` uses<br>
:func:`~sklearn.metrics.RocCurveDisplay.plot` to plot the SVC ROC curve<br>
without recomputing the values of the roc curve itself. Furthermore, we<br>
pass `alpha=0.8` to the plot functions to adjust the alpha values of the<br>
curves.

```python
rfc = RandomForestClassifier(n_estimators=10, random_state=42)
rfc.fit(X_train, y_train)
ax = plt.gca()
rfc_disp = plot_roc_curve(rfc, X_test, y_test, ax=ax, alpha=0.8)
svc_disp.plot(ax=ax, alpha=0.8)
plt.show()
```
