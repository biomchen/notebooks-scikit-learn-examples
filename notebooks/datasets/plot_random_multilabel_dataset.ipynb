Plot randomly generated multilabel dataset<br>
==============================================<br>
This illustrates the :func:`~sklearn.datasets.make_multilabel_classification`<br>
dataset generator. Each sample consists of counts of two features (up to 50 in<br>
total), which are differently distributed in each of two classes.<br>
Points are labeled as follows, where Y means the class is present:<br>
    =====  =====  =====  ======<br>
      1      2      3    Color<br>
    =====  =====  =====  ======<br>
      Y      N      N    Red<br>
      N      Y      N    Blue<br>
      N      N      Y    Yellow<br>
      Y      Y      N    Purple<br>
      Y      N      Y    Orange<br>
      Y      Y      N    Green<br>
      Y      Y      Y    Brown<br>
    =====  =====  =====  ======<br>
A star marks the expected sample for each class; its size reflects the<br>
probability of selecting that class label.<br>
The left and right examples highlight the ``n_labels`` parameter:<br>
more of the samples in the right plot have 2 or 3 labels.<br>
Note that this two-dimensional example is very degenerate:<br>
generally the number of features would be much greater than the<br>
"document length", while here we have much larger documents than vocabulary.<br>
Similarly, with ``n_classes > n_features``, it is much less likely that a<br>
feature distinguishes a particular class.<br>


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_multilabel_classification as make_ml_clf
```

```python
print(__doc__)
```

```python
COLORS = np.array(['!',
                   '#FF3333',  # red
                   '#0198E1',  # blue
                   '#BF5FFF',  # purple
                   '#FCD116',  # yellow
                   '#FF7216',  # orange
                   '#4DBD33',  # green
                   '#87421F'   # brown
                   ])
```

Use same random seed for multiple calls to make_multilabel_classification to<br>
ensure same distributions

```python
RANDOM_SEED = np.random.randint(2 ** 10)
```

```python
def plot_2d(ax, n_labels=1, n_classes=3, length=50):
    X, Y, p_c, p_w_c = make_ml_clf(n_samples=150, n_features=2,
                                   n_classes=n_classes, n_labels=n_labels,
                                   length=length, allow_unlabeled=False,
                                   return_distributions=True,
                                   random_state=RANDOM_SEED)
    ax.scatter(X[:, 0], X[:, 1], color=COLORS.take((Y * [1, 2, 4]
                                                    ).sum(axis=1)),
               marker='.')
    ax.scatter(p_w_c[0] * length, p_w_c[1] * length,
               marker='*', linewidth=.5, edgecolor='black',
               s=20 + 1500 * p_c ** 2,
               color=COLORS.take([1, 2, 4]))
    ax.set_xlabel('Feature 0 count')
    return p_c, p_w_c
```

```python
_, (ax1, ax2) = plt.subplots(1, 2, sharex='row', sharey='row', figsize=(8, 4))
plt.subplots_adjust(bottom=.15)
```

```python
p_c, p_w_c = plot_2d(ax1, n_labels=1)
ax1.set_title('n_labels=1, length=50')
ax1.set_ylabel('Feature 1 count')
```

```python
plot_2d(ax2, n_labels=3)
ax2.set_title('n_labels=3, length=50')
ax2.set_xlim(left=0, auto=True)
ax2.set_ylim(bottom=0, auto=True)
```

```python
plt.show()
```

```python
print('The data was generated from (random_state=%d):' % RANDOM_SEED)
print('Class', 'P(C)', 'P(w0|C)', 'P(w1|C)', sep='\t')
for k, p, p_w in zip(['red', 'blue', 'yellow'], p_c, p_w_c.T):
    print('%s\t%0.2f\t%0.2f\t%0.2f' % (k, p, p_w[0], p_w[1]))
```
