The Digit Dataset<br>
=========================================================<br>
This dataset is made up of 1797 8x8 images. Each image,<br>
like the one shown below, is of a hand-written digit.<br>
In order to utilize an 8x8 figure like this, we'd have to<br>
first transform it into a feature vector with length 64.<br>
See `here<br>
<https://archive.ics.uci.edu/ml/datasets/Pen-Based+Recognition+of+Handwritten+Digits>`_<br>
for more information about this dataset.<br>
