SVM-Anova: SVM with univariate feature selection<br>
=================================================<br>
This example shows how to perform univariate feature selection before running a<br>
SVC (support vector classifier) to improve the classification scores. We use<br>
the iris dataset (4 features) and add 36 non-informative features. We can find<br>
that our model achieves best performance when we select around 10% of features.<br>


```python
print(__doc__)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectPercentile, chi2
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

#############################################################################<br>
Import some data to play with

```python
X, y = load_iris(return_X_y=True)
# Add non-informative features
np.random.seed(0)
X = np.hstack((X, 2 * np.random.random((X.shape[0], 36))))
```

#############################################################################<br>
Create a feature-selection transform, a scaler and an instance of SVM that we<br>
combine together to have an full-blown estimator

```python
clf = Pipeline([('anova', SelectPercentile(chi2)),
                ('scaler', StandardScaler()),
                ('svc', SVC(gamma="auto"))])
```

#############################################################################<br>
Plot the cross-validation score as a function of percentile of features

```python
score_means = list()
score_stds = list()
percentiles = (1, 3, 6, 10, 15, 20, 30, 40, 60, 80, 100)
```

```python
for percentile in percentiles:
    clf.set_params(anova__percentile=percentile)
    this_scores = cross_val_score(clf, X, y)
    score_means.append(this_scores.mean())
    score_stds.append(this_scores.std())
```

```python
plt.errorbar(percentiles, score_means, np.array(score_stds))
plt.title(
    'Performance of the SVM-Anova varying the percentile of features selected')
plt.xticks(np.linspace(0, 100, 11, endpoint=True))
plt.xlabel('Percentile')
plt.ylabel('Accuracy Score')
plt.axis('tight')
plt.show()
```
